```python
# how to read missing data
```

```python
df=pd.read_csv('weather1_data.csv')
```

```python
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
# nan replace with onter values
```

```python
new_df=df.fillna({
    "temperature":0,
        "windspeed":0,
        "event":"noevent",

})
new_df
```

```python
new_df = df.fillna(method="ffill") # take the before vales
new_df
```

```python
new_df = df.fillna(method="bfill")# back
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)# only change one 
new_df
```

```python
new_df = df.interpolate()# avge of b and ofter
new_df
```

```python
new_df = df.dropna()
new_df
# missing data
```
